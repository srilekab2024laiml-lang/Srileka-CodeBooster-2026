```python
def ask_llm(user_message, system_message="you are a helpful assistant,",
            temperature=0.7, max_token=500):
  response = clint.chat.completions.create(
    model=MODEL,
    messages=[
      {"role": "system", "content": system_message},
      {"role": "user", "content": user_message}
    ],
  temperature=temperature,
    max_tokens=max_token,
    stream=False
  )
  return response.choices[0].message.content

test_response = ask_llm(
    "What is the meaning of Srileka"
  )
print("=== LLM Response ===")
print(test_response)
```

```python
# This section demonstrates how to tailor LLM output for coding students
# by leveraging system_message and managing tokens.

# 1. Define a clear system message to set the LLM's persona.
#    This guides the LLM to respond as if it's talking to a coding student.
system_message_for_students = "You are an experienced software engineering instructor. Explain technical concepts clearly and concisely, using code-relevant terminology where appropriate, and assume the audience has a basic understanding of programming. Keep answers practical."

# 2. Craft a user message that asks a relevant question.
user_question_for_llm = "Explain the concept of 'API endpoint' for a coding student. Include an example of a common HTTP method."

# 3. Call the ask_llm function.
#    We'll also set a max_token to demonstrate controlling output length,
#    which is crucial for cost and conciseness in real-world applications.
llm_response_for_students = ask_llm(
    user_message=user_question_for_llm,
    system_message=system_message_for_students,
    temperature=0.5, # Keep it a bit deterministic for technical explanations
    max_token=150 # Limit the response length to around 100-150 words
)

print("=== LLM Output for Coding Students ===")
print(f"System Role: '{system_message_for_students}'")
print(f"User Query: '{user_question_for_llm}'")
print("---------------------------------------")
print(llm_response_for_students)
print("\n--- Token Explanation ---")
print("Notice how the 'system_message' shapes the response's tone and content. \
      The 'max_token' parameter directly controls the maximum length of the output, \
      which is important for managing API costs and ensuring conciseness. \
      Roughly, 1 word is 1-2 tokens, so 150 tokens means about 75-150 words.")
```

```python
few_shot_user_message = """
Convert employee text to a JSON object. Provide ONLY the JSON object, without any additional text, explanations, or markdown formatting.

Here are examples:
Input: Ramesh Kumar, 4500, Mumbai
Output: {"name": "Ramesh Kumar", "salary": 4500, "city": "Mumbai"}

Input: Tanvi Mehta, 5500, Bangalore
Output: {"name": "Tanvi Mehta", "salary": 5500, "city": "Bangalore"}

Now, convert the following:
Input: Ravi Patel, 6000, Delhi
Output:"""

few_shot_response = ask_llm(
    user_message=few_shot_user_message,
    system_message="You are an expert at converting data into JSON format.", # Simpler system message, as strictness is in user_message
    temperature=0.0
)
print('few-shot Result:')
print(few_shot_response)
print()

try:
  parsed =json.loads(few_shot_response.strip())
  print('Successfully parsed JSON')
  print(f'Name: {parsed["name"]}')
  print(f'Salary: {parsed["salary"]}')
  print(f'City: {parsed["city"]}')
except Exception as e:
  print(f'Parsed failed - model added extra text or invalid JSON: {e}')
  print('Solution: The model needs to be more strictly instructed to return ONLY JSON, without markdown or extra text.')
```

```python
name_classification_user_message = """
Classify the given name as 'Girl' or 'Boy'. Provide ONLY the classification (Girl or Boy).

Here are examples:
Input: Ramesh Kumar
Output: Boy

Input: Tanvi Mehta
Output: Girl

Input: Ravi Patel
Output: Boy

Input: Ananya Sharma
Output: Girl

Input: Ravi Patel
Output:"""

name_classification_response = ask_llm(
    user_message=name_classification_user_message,
    system_message="You are an expert in classifying names by gender.",
    temperature=0.0, # Keep it deterministic for classification
    max_token=5 # A very small max_token is sufficient for 'Girl' or 'Boy'
)

print('--- Name Classification Result ---')
print(f"Input Name: Ravi patel")
print(f"Classification: {name_classification_response.strip()}")

print("\n--- Explanation ---")
print("By providing clear examples of names and their classifications, we've guided the LLM \n to perform this specific task. The 'system_message' reinforces its role, \n      and 'temperature=0.0' ensures consistent output. 'max_token' is set low \n      because the expected output is very short.")
```
