```python
cs_students = (df['department'] == 'Computer Science')
print(f"No of Computer Science students: {len(df[cs_students])}")
print("\nComputer Science students:")
print(df[cs_students][['name', 'math_score', 'programming_score', 'attendance_percentage']])
```

```python
above_90 = df[df['attendance_percentage'] > 90]
print("Students with attendance above 90%:\n")
print(above_90[['name', 'department', 'attendance_percentage']])
```

```python
top_performers = df.sort_values(by='programming_score', ascending=False)

print(top_performers[['name', 'programming_score']].head(10))
```

```python
df['total_score'] = df['math_score'] + df['programming_score']

above_70 = df[df['total_score'] > 70]

print("Number of students above 70:", len(above_70))

top_performers = df.sort_values(
    by='total_score',
    ascending=False
)

print("\nTop Overall Performers:\n")

print(top_performers[['name', 'department', 'total_score']].head(10))
```

```python
df['total_score'] = df['math_score'] + df['programming_score']

top_performers = df.sort_values(
    by='total_score',
    ascending=False
)

print("Top Overall Performers:\n")

print(top_performers[['name', 'department', 'total_score']].head(10))
```

```python
avg_math_score_per_department = df.groupby('department')['math_score'].mean()
print("Average Math Score per Department:")
print(avg_math_score_per_department)
```

```python
dept_avg = df.groupby('department')[['math_score', 'programming_score']].mean()
print("Average total score:")
print(dept_avg)
```

```python
dept_avg = df.groupby('department')[['math_score', 'programming_score']].mean()
dept_avg['total_score'] = (dept_avg['math_score'] + dept_avg['programming_score'])

dept_avg = dept_avg.sort_values( by='total_score',ascending=False)
print(dept_avg)
```
