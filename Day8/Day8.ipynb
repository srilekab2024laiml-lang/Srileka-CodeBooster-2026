```python
def retrieve_notes(question, collection, top_k=5):
    question_embedding = embedding_model.encode(question)
    results = collection.query(
        query_embeddings=[question_embedding],
        n_results=top_k
    )
    return results

print("Retrieval function `retrieve_notes` defined.")
```

```python
def generate_rag_answer(question, context):
    system_prompt = "You are a helpful acadamic assistant for engineering students."
    user_prompt = f"""Context from Knowledge Base:

{context}

{question}
"""

    response = groq_client.chat.completions.create(
        model="llama-3.1-8b instant", # Added comma here
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0.1,
        max_tokens=500
    )

    answer = response.choices[0].message.content
    return answer
```
